```python
# your code here

#Grouping by customer ID and Country in a new df

customers = orders.groupby(['CustomerID', 'Country']).agg({'amount_spent':'sum'})
customers.head()
```

```python
labels = ["very low", "low", "medium", "high", "very high"]
```

```python
customers["spend_bins"] = pd.qcut(customers['amount_spent'], 5, labels = labels)
customers
```

```python
customers["spend_bins"].value_counts()    

#I managed to separate them into bins but I don't how to make it so the top bin is only above the 95%.
#In my example, VIP customers are the Very High bin and the preferred customers are the High bin
```

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here

vips = customers[customers["spend_bins"] == "very high"]
vips
```

```python
vips = vips.reset_index()
```

```python
vips["Country"].value_counts()   #United Kingdom is the country with most VIP customers
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here

vipspref = customers[(customers["spend_bins"] == "very high") | (customers["spend_bins"] == "high")]
vipspref = vipspref.reset_index()
vipspref
```

```python
vipspref["Country"].value_counts() #UK also leads on VIPs and Preferred customers
```
